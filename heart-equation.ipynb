```python
# Programa em Python para desenhar um lindo coração

# importando bibliotecas
import numpy as np
from matplotlib import pyplot as plt

# Criando 100 dados igualmente espaçados no intervalo de 0 a 2*pi
theta = np.linspace(0, 2 * np.pi, 100)

# Gerando x e y
x = 16 * ( np.sin(theta) ** 3 )
y = 13 * np.cos(theta) - 5* np.cos(2*theta) - 2 * np.cos(3*theta) - np.cos(4*theta)

# Desenhando
plt.plot(x, y, 'red')
plt.title('Coração', color='pink')
plt.show()
```
